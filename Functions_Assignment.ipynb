Q6  Explain the concept of generators in Python and how they are defined.

Ans6. A generator is a special type of iterator in Python.

Instead of storing all values in memory, it generates values on the fly when you need them.

This makes generators memory efficient and useful for handling large or infinite sequences.

```python
# Ans 6.

def countdown(n):
    while n > 0:
        yield n
        n -= 1

for num in countdown(5):
    print(num)
```

Q7.  What are the advantages of using generators over regular functions?

Ans7. Generators are better than regular functions when you need memory efficiency, lazy evaluation, handling of large or infinite data, and cleaner pipeline-style code.

```python
# Ans7
# Produces numbers (Lazy + Infinite capability)
def numbers():
    n = 1
    while True:       # Infinite sequence possible
        yield n
        n += 1

# Squares the numbers (Pipeline processing)
def squares(seq):
    for num in seq:
        yield num * num

# Filters even squares (Pipeline step 2)
def even_only(seq):
    for num in seq:
        if num % 2 == 0:
            yield num

nums = numbers()          # Infinite generator
sq = squares(nums)        # Pipeline: numbers -> squares
evens = even_only(sq)     # Pipeline: filter evens

#  first 5 even squares lazily
for _ in range(5):
    print(next(evens))
```

Q10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

Ans10. 1.  a) map

Purpose:-

Applies a given function to each element of an iterable.

Produces a new iterable with the transformed elements.

b) filter

Purpose:-

Applies a function (that returns True/False) to each element.

Keeps only the elements for which the function returns True.

c) 3. reduce

Imported from functools (from functools import reduce).

Purpose

Applies a function cumulatively to the elements of an iterable, reducing it to a single value.


```python
# Ans10

from functools import reduce


nums = [1, 2, 3, 4, 5, 6]

# 1. map() → Square each number
squares = map(lambda x: x*x, nums)
# → [1, 4, 9, 16, 25, 36]

# 2. filter() → Keep only even squares
even_squares = filter(lambda x: x % 2 == 0, squares)
# → [4, 16, 36]

# 3. reduce() → Sum all even squares
result = reduce(lambda x, y: x + y, even_squares)


print(result)
```

```python
#Q1.  Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.

def sum_of_even_numbers(numbers):
    sum = 0
    for num in numbers:
        if num % 2 == 0:
            sum += num
    return sum


numbers = [1, 2, 3, 4, ]
sum_of_even_numbers(numbers)

```

```python
#Q4. Write a Python function that checks if a given number is prime or not from 1 to 200.

def is_prime(n):
    if n <= 1:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False

    for i in range(3, int(n**0.5) + 1, 2):
        if n % i == 0:
            return False
    return True

is_prime(19)
```

```python
#Q5.  Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms

class Fibonacci:
    def __init__(self, n_terms):
        self.n_terms = n_terms
        self.count = 0
        self.a, self.b = 0, 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.count < self.n_terms:
            if self.count == 0:
                self.count += 1
                return 0
            elif self.count == 1:
                self.count += 1
                return 1
            else:
                self.a, self.b = self.b, self.a + self.b
                self.count += 1
                return self.b
        else:
            raise StopIteration

a = Fibonacci(5)

for num in a:
    print(num, end=" ")



```

```python
#Q6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

def powers_of_two(limit):
    for exp in range(limit + 1):
        yield 2 ** exp

for val in powers_of_two(5):
    print(val, end=" ")

```
